### Task 3. Data visualization

We’ve assessed the data, and decided on which data variables are most applicable. It’s time to plot the visualization(s).


#### Boxplots

Perform a check for outliers on relevant columns such as trip distance and trip duration, but also on total amount paid and tip amount by vendor to see if there are any interesting trends.

**Note:** We will also convert the date columns to datetime in order to derive total trip duration.  

```python
# Convert to datetime format the columns tpep_pickup_datetime and tpep_dropoff_datetime
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime']=pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# Create a boxplot of the column trip_distance to find outliers
plt.figure(figsize=(7,2))
plt.title('trip_distance')
sns.boxplot(data=None, x=df['trip_distance'], fliersize=1);
```

```python
# Create a histogram of the column trip_distance to find outliers
plt.figure(figsize=(10,5))
sns.histplot(df['trip_distance'], bins=range(0,26,1))
plt.title('Trip distance histogram');
```

```python
# Create a boxplot with the total amount variable to find outliers
plt.figure(figsize=(7,2))
plt.title('total_amount')
sns.boxplot(x=df['total_amount'], fliersize=1);
```

```python
# Create a histogram with the total amount variable to find outliers
plt.figure(figsize=(12,6))
ax = sns.histplot(df['total_amount'], bins=range(-10,101,5))
ax.set_xticks(range(-10,101,5))
ax.set_xticklabels(range(-10,101,5))
plt.title('Total amount histogram');
```

```python
# Create a boxplot with the tip amount variable to find outliers
plt.figure(figsize=(7,2))
plt.title('tip_amount')
sns.boxplot(x=df['tip_amount'], fliersize=1);
```

```python
# Create a histogram with the tip amount variable to find outliers
plt.figure(figsize=(12,6))
ax = sns.histplot(df['tip_amount'], bins=range(0,21,1))
ax.set_xticks(range(0,21,2))
ax.set_xticklabels(range(0,21,2))
plt.title('Tip amount histogram');
```

```python
# Create a stacked histplot with the tip amount by vendor id to find outliers
plt.figure(figsize=(12,7))
ax = sns.histplot(data=df, x='tip_amount', bins=range(0,21,1), 
                  hue='VendorID', 
                  multiple='stack',
                  palette='pastel')
ax.set_xticks(range(0,21,1))
ax.set_xticklabels(range(0,21,1))
plt.title('Tip amount by vendor histogram');
```

```python
# Create a stacked histplot with the tip amount above 10 by vendor to display this low metric section that is barely visible in the previous graph
tips_over_ten = df[df['tip_amount'] > 10]
plt.figure(figsize=(12,7))
ax = sns.histplot(data=tips_over_ten, x='tip_amount', bins=range(10,21,1), 
                  hue='VendorID', 
                  multiple='stack',
                  palette='pastel')
ax.set_xticks(range(10,21,1))
ax.set_xticklabels(range(10,21,1))
plt.title('Tip amount by vendor histogram');
```

#### Scatter plot

#### We notice that there is some trips in the data with 0 miles distance and still charged a fare. We will remove those trips from the data to avoid skewing the data. Therefore we will remove those trips with costs associated, but with trip distance = 0.*

```python
df_2 = df['trip_distance'].loc[~(df==0).all(axis=1)]
```

```python
# Create a scatterplot with the total amount by trip distance with the filtered dataframe created previously
sns.scatterplot(x=df['total_amount'], y=df_2)
```

```python
# Analyze the passenger count variable per trip
df['passenger_count'].unique()
```

```python
# Create a barplot with the passenger count distribution in the dataframe, use pastel palette to stay in the theme
plt.figure(figsize=(10,5))
ax = sns.countplot(x=df['passenger_count'], palette='pastel')
ax.set_xticklabels(ax.get_xticklabels())
plt.title('Passenger count distribution');
```

```python
# Group by the passenger count and the mean tip amount, then create an sns barplot with the results, use pastel palette
df.groupby('passenger_count')['tip_amount'].mean()
plt.figure(figsize=(10,5))
ax = sns.barplot(x=df['passenger_count'], y=df['tip_amount'], palette='pastel')
ax.set_xticklabels(ax.get_xticklabels())
plt.title('Tip amount by passenger count');
```

```python
# We will create the mean_tips_by_passenger_count = one, two and three, it does not make sense to focus on 0 passenger count.
one_passenger = df[df['passenger_count']==1]['tip_amount'].mean()
two_passenger = df[df['passenger_count']==2]['tip_amount'].mean()
three_passenger = df[df['passenger_count']==3]['tip_amount'].mean()
```

```python
# Group by the mean drop off location id and the trip distance, then create an sns barplot with the results, use pastel palette
# Filter only on payment type 1 and 2 to focus on credit card and cash
df.groupby('DOLocationID')['trip_distance'].mean()
sorted_df = df[df['payment_type'].isin([1,2])]
plt.figure(figsize=(10,5))
ax = sns.barplot(x=df['DOLocationID'], y=df['trip_distance'], palette='pastel')
ax.set_xticklabels(ax.get_xticklabels())
plt.title('Trip distance by drop off location');
```

### Task 4a. Results and evaluation

Having built visualizations in Tableau and in Python, I have discovered the following key insights:

* The highest distribution of trip distances are below 5 miles, but there are outliers all the way out to 35 miles. There are no missing values. 
* For the entire year, Vendor 2 was able to get more customers and trips than Vendor 1.
* Vendor 2 made the longest trip, which was 33.96 miles, and the yearly sum charged was 206, 000 dollars. The total distance driven in a year was approximately 34, 000 miles.
* 71% of the trips had just one passenger, while 29% had two passengers or more.
* The average tip was $1.35, paid primarily by card (32% of passengers pay by cash), and there was no significant difference in tip amounts between trips with one passenger or more.
* In 22K trips, 24% of the drop-off locations ID range between area 220 and 240, which corresponds to the highest drop-off locations evenly distributed between vendors 1 and 2.
